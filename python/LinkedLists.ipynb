# Linked Lists
A linked list is a chain of nodes
Each node stores:
    1. a value
    2. a pointer/reference to next node (address to next value)
Example: 1 -> 2 -> 3 -> 4 -> None
the tail points to None

Lets code a Node first 

````python
class LinkNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next
    
    def __repr__(self):
        return f'ListNode with value of {self.value}'

        
````

now lets create a few nodes manually

````python
a = ListNode(1)
b = ListNode(2)
c = ListNode(3)

# now lets connect them by assigning their nexts a value ie the next element

a.next = b
b.next = c
# by default c.next is None

#if head is a you can technically get the whole linked list with just the head 
#because every next element will have the nexts value 
# head = a
# print(head)
print(b)
````

now we have the logic lets implement some functions with the logic and the linked list elements


````python
def build_linked_list(List):
    
    #taking care of [] empty list input 
    if not List:
        return None

    # building a linked list from a list
    head = ListNode(List[0])
    current_value = head

    for list_element in List[1:]: #excluded the head slicing from the index 1 
        current_value.next = ListNode(list_element) #now make the next item/element also a ListNode type
        current_value = current_value.next #and next and next...
    
    return head

#building a python list from Linked list
def linked_list_to_list(head):
    List = []
    curr = head

    while curr: # when current will be None i.e at the last element's next loop will stop
        List.append(curr.value) 
        curr = curr.next
    
    return List

#printing the linked list
def print_linked_list(head: ListNode): #the argument which you will pass head will be of type ListNode
    if not head:
        print('None')
        return

    linked_list_values = []
    curr = head

    while curr:
        linked_list_values.append(str(curr.value)) # we could have used linked_list_to_list() but we need str list
        curr = curr.next
    
    print('->'.join(linked_list_values) + "-> None")


    
    #None input
    

````

````python
head = build_linked_list([1, 2, 3, 4]) #building linked list returns a head 
print_linked_list(head)
linked_list_to_list(head)

# earlier we manually assigned a.next = b & b.next = c
# in short if you get the head you get the whole linked list
print_linked_list(a)
````

## operations on linked lists
now lets learn operations like insert at begining, insert at end, and insert at a given value
these are basics of pointers and pointer manipulations


````python
#insert at beginning
def insert_at_beginning(head: ListNode, value: int):
    new_node = ListNode(value)
    new_node.next = head
    return new_node

#inserting at the end
def insert_at_end(head: ListNode, value: int):
    new_node = ListNode(value)

    if not head:
        return new_node
    
    curr = head
    while curr.next: # we want curr.next to be None i. e  we will reach the end 
        curr = curr.next
    
    curr.next = new_node
    return head

# inserting at a given value 
````
